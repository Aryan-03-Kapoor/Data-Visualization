```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
data = pd.read_csv('StudentsPerformance.csv')
data
```

```python
a=data[data.lunch== 'standard']
b=data[data.lunch== 'free/reduced']
```

```python
M_ms=a["math score"].mean()
M_rs=a["reading score"].mean()
M_ws=a["writing score"].mean()
```

```python
F_ms=b["math score"].mean()
F_rs=b["reading score"].mean()
F_ws=b["reading score"].mean()
```

```python
plt.title("Average On Basis of Lunch")
plt.ylabel("Marks of Maths")
plt.xlabel("Lunch")
plt.bar(a.lunch,M_ms,width=0.4)
plt.bar(b.lunch,F_ms,width=0.)
plt.legend(["Standatrd","Free/Reduced"])
```

```python
plt.title("Average On Basis of Lunch")
plt.ylabel("Marks of Reading")
plt.xlabel("Lunch")
plt.bar(a.lunch,M_rs,width=0.4)
plt.bar(b.lunch,F_rs,width=0.4)
plt.legend(["Standatrd","Free/Reduced"])
```

```python
plt.title("Average On Basis of Lunch")
plt.ylabel("Marks of Writing")
plt.xlabel("Lunch")
plt.bar(a.lunch,M_ws,width=0.4)
plt.bar(b.lunch,F_ws,width=0.4)
plt.legend(["Standatrd","Free/Reduced"])
```

```python
A=data[data["race/ethnicity"]== "group A"]
B=data[data['race/ethnicity']=="group B"]
C=data[data["race/ethnicity"]== "group C"]
D=data[data['race/ethnicity']=="group D"]
E=data[data["race/ethnicity"]== "group E"]
```

```python
M_ms=A["math score"].mean()
M_rs=A["reading score"].mean()
M_ws=A["writing score"].mean()

F_ms=B["math score"].mean()
F_rs=B["reading score"].mean()
F_ws=B["writing score"].mean()

N_ms=C["math score"].mean()
N_rs=C["reading score"].mean()
N_ws=C["writing score"].mean()

W_ms=D["math score"].mean()
W_rs=D["reading score"].mean()
W_ws=D["writing score"].mean()

X_ms=E["math score"].mean()
X_rs=E["reading score"].mean()
X_ws=E["writing score"].mean()
```

```python
plt.figure(figsize=(13,13))
plt.title("Average On Basis of Group")
plt.ylabel("Marks of Maths")
plt.xlabel("Group")
plt.bar(A["race/ethnicity"],M_ms,width=0.4)
plt.bar(B["race/ethnicity"],F_ms,width=0.4)
plt.bar(C["race/ethnicity"],N_ms,width=0.4)
plt.bar(D["race/ethnicity"],W_ms,width=0.4)
plt.bar(E["race/ethnicity"],X_ms,width=0.4)
plt.legend(["Group A","Group B","Group C","Group D","Group E"])
```

```python
plt.figure(figsize=(13,15))
plt.title("Average On Basis of Group")
plt.ylabel("Marks of Reading")
plt.xlabel("Group")
plt.bar(A["race/ethnicity"],M_rs,width=0.4)
plt.bar(B["race/ethnicity"],F_rs,width=0.4)
plt.bar(C["race/ethnicity"],N_rs,width=0.4)
plt.bar(D["race/ethnicity"],W_rs,width=0.4)
plt.bar(E["race/ethnicity"],X_rs,width=0.4)
plt.legend(["Group A","Group B","Group C","Group D","Group E"])
```

```python
plt.figure(figsize=(13,13))
plt.title("Average On Basis of Group")
plt.ylabel("Marks of Writing")
plt.xlabel("Group")
plt.bar(A["race/ethnicity"],M_ws,width=0.4)
plt.bar(B["race/ethnicity"],F_ws,width=0.4)
plt.bar(C["race/ethnicity"],N_ws,width=0.4)
plt.bar(D["race/ethnicity"],W_ws,width=0.4)
plt.bar(E["race/ethnicity"],X_ws,width=0.4)
plt.legend(["Group A","Group B","Group C","Group D","Group E"])
```
